### FUNCTION TO PLOT THE LOSS CURVES

```python

def plot_loss_curves(history):
  """
  Return separate loss curves for training and validation data
  """
  loss = history.history["loss"]
  val_loss = history.history["val_loss"]

  accuracy =history.history["accuracy"]
  val_accuracy= history.history["val_accuracy"]

  epochs= range(len(history.history["loss"]))

  plt.plot(epochs , loss , label = "training data")
  plt.plot(epochs , val_loss , label ="val_accuracy")
  plt.title("loss")
  plt.xlabel("epochs")
  plt.legend()

  plt.figure()
  plt.plot(epochs , accuracy , label = "training_accuracy")
  plt.plot(epochs , val_accuracy , label ="val_accuracy")
  plt.title("accuracy")
  plt.xlabel("epochs")
  plt.legend()
```

### FUNCTION TO MAKE CONFUSION MATRIX

```python
import itertools
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix

def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15, norm=False, savefig=False):
  """Makes a labelled confusion matrix comparing predictions and ground truth labels.

  If classes is passed, confusion matrix will be labelled, if not, integer class values
  will be used.

  Args:
    y_true: Array of truth labels (must be same shape as y_pred).
    y_pred: Array of predicted labels (must be same shape as y_true).
    classes: Array of class labels (e.g. string form). If `None`, integer labels are used.
    figsize: Size of output figure (default=(10, 10)).
    text_size: Size of output figure text (default=15).
    norm: normalize values or not (default=False).
    savefig: save confusion matrix to file (default=False).

  Returns:
    A labelled confusion matrix plot comparing y_true and y_pred.

  Example usage:
    make_confusion_matrix(y_true=test_labels, # ground truth test labels
                          y_pred=y_preds, # predicted labels
                          classes=class_names, # array of class label names
                          figsize=(15, 15),
                          text_size=10)
  """
  # Create the confustion matrix
  cm = confusion_matrix(y_true, y_pred)
  cm_norm = cm.astype("float") / cm.sum(axis=1)[:, np.newaxis] # normalize it
  n_classes = cm.shape[0] # find the number of classes we're dealing with

  # Plot the figure and make it pretty
  fig, ax = plt.subplots(figsize=figsize)
  cax = ax.matshow(cm, cmap=plt.cm.Blues) # colors will represent how 'correct' a class is, darker == better
  fig.colorbar(cax)

  # Are there a list of classes?
  labels = classes

  # Label the axes
  ax.set(title="Confusion Matrix",
         xlabel="Predicted label",
         ylabel="True label",
         xticks=np.arange(n_classes), # create enough axis slots for each class
         yticks=np.arange(n_classes),
         xticklabels=labels, # axes will labeled with class names (if they exist) or ints
         yticklabels=labels)

  # Make x-axis labels appear on bottom
  ax.xaxis.set_label_position("bottom")
  ax.xaxis.tick_bottom()
  ax.tick_params(axis='x', labelrotation = 75)
  # Set the threshold for different colors
  threshold = (cm.max() + cm.min()) / 2.

  # Plot the text on each cell
  for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    if norm:
      plt.text(j, i, f"{cm[i, j]} ({cm_norm[i, j]*100:.1f}%)",
              horizontalalignment="center",
              color="white" if cm[i, j] > threshold else "black",
              size=text_size)
    else:
      plt.text(j, i, f"{cm[i, j]}",
              horizontalalignment="center",
              color="white" if cm[i, j] > threshold else "black",
              size=text_size)

  # Save the figure to the current working directory
  if savefig:
    fig.savefig("confusion_matrix.png")
```

# Extracting features of a song

# variables and their significance:
`y`: `audio time series`

`sr`: `sampling rate of y`

```python

# Function to extract audio features
def extract_audio_features(audio_file):
    # Load the audio file
    y, sr = librosa.load(audio_file)   ##

    # Extract features
    onset_strength = np.mean(librosa.onset.onset_strength(y=y, sr=sr))
    chroma_stft = np.mean(librosa.feature.chroma_stft(y=y, sr=sr))
    chroma_cqt = np.mean(librosa.feature.chroma_cqt(y=y, sr=sr))
    chroma_cens = np.mean(librosa.feature.chroma_cens(y=y, sr=sr))
    melspectrogram = np.mean(librosa.feature.melspectrogram(y=y, sr=sr))
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr))
    spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
    spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
    spectral_contrast = np.mean(librosa.feature.spectral_contrast(y=y, sr=sr))
    spectral_rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))
    tonnetz = np.mean(librosa.feature.tonnetz(y=y, sr=sr))
    zero_crossing_rate = np.mean(librosa.feature.zero_crossing_rate(y=y))


    return {
        'onset_strength': onset_strength,
        'chroma_stft': chroma_stft,
        'chroma_cqt': chroma_cqt,
        'chroma_cens': chroma_cens,
        'melspectrogram': melspectrogram,
        'mfcc': mfcc,
        'spectral_centroid': spectral_centroid,
        'spectral_bandwidth': spectral_bandwidth,
        'spectral_contrast': spectral_contrast,
        'spectral_rolloff': spectral_rolloff,
        'tonnetz': tonnetz,
        'zero_crossing_rate': zero_crossing_rate
    }

# Example usage:
```

### Condition should be subject to change depending on different cases

```python
import random
import math

# Example usage:

def num_recommendations(class_ratios_dict):
    """
    Assign numerical values to classes based on their ratio values.

    Parameters:
    - class_ratios_dict: A dictionary with class labels as keys and ratio values as values.

    Returns:
    - A dictionary with class labels as keys and numerical values based on the ratio values.
    """
    class_value_dict = {}  # Dictionary to store class values

    for class_label, ratio in class_ratios_dict.items():
        if ratio < 0.05:
            class_value = 0
        elif 0.05 <= ratio < 0.15:
            class_value = 2
        elif 0.15 <= ratio < 0.3:
            class_value = 3
        elif 0.3 <= ratio < 0.5:
            class_value = 4
        elif 0.5 <= ratio < 0.7:
            class_value = 5
        else:
            class_value = 6

        class_value_dict[class_label] = class_value

    return class_value_dict



# Assuming 'df' is your DataFrame with the last column named 'class_column'
# You can replace 'class_column' with the actual name of your last column

# Count the occurrences of each class in the last column
class_counts = sample_user_playlist_data['Class'].value_counts()

# Convert the result to a dictionary
class_counts_dict = class_counts.to_dict()
print(class_counts_dict)

total_count = len(sample_user_playlist_data)

class_ratios_dict = (class_counts / total_count).to_dict()

print("Ratios in user playlist:",class_ratios_dict)

num_songs_rec = num_recommendations(class_ratios_dict)
print("Number of songs to recommend from each class:",num_songs_rec)
```

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors

def k_nearest_neighbors(input_point, data, k):
    """
    Find the k-nearest neighbors to the input point in a dataset.

    Parameters:
    - input_point: A 1D NumPy array representing the point in 12-dimensional space.
    - data: A 2D NumPy array representing the dataset where each row is a data point.
    - k: The number of nearest neighbors to find.

    Returns:
    - An array of indices of the k-nearest neighbors in the dataset.
    """

    raw_data = np.array(data.iloc[:,1:13])

    # Initialize the NearestNeighbors model
    nbrs = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(raw_data)

    # Find the indices of the k-nearest neighbors
    distances, indices = nbrs.kneighbors([input_point])

    songs_df = []
    for i in range(indices[0].size):
      songs_df.append(data.iloc[indices[0][i],0])

    return songs_df

final_recommendations = []

for index, row in means_df.iterrows():
    # 'index' is the row index, 'row' is a Series representing the current row
    class_name = row['Class']
    num_recommendations = row['NumRecommendations']

    # Assuming you have separate DataFrames for each class like 'class1_df', 'class2_df', etc.
    #class_df = globals()[class_name + '_df']
    class_df = training_data.loc[training_data.loc[: ,'Class'] == class_name]
    # Convert the row to a NumPy array
    input_point = row.iloc[0:12].to_numpy()

    # Find the k-nearest neighbors for the input point within the class DataFrame
    nearest_songs = k_nearest_neighbors(input_point, class_df, num_recommendations)

    final_recommendations.append(nearest_songs)


```
